```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,ElementNotInteractableException

#Importing requests
import requests

# importing regex
import re
from selenium.webdriver.common.by import By
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)

# Opening the wikipedia
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
name=[]
try:
    names=driver.find_elements(By.XPATH,"//*[@id='mw-content-text']/div[1]/table[2]/tbody/tr/td[2]")
    for i in names:
        name.append(i.text)
except NoSuchElementException:#handling no such element exception
    name.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    name.append('No details available')
    
time.sleep(2) 
#scraping artist/uploader of the video
artist=[]
try:
    artists=driver.find_elements(By.XPATH,"//*[@id='mw-content-text']/div[1]/table[2]/tbody/tr/td[3]")
    for i in artists:
        artist.append(i.text)
except NoSuchElementException:#handling no such element exception
    artist.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    artist.append('No details available')
    
time.sleep(2) 
#scraping views of the video
views=[]
try:
    views_list=driver.find_elements(By.XPATH,"//*[@id='mw-content-text']/div[1]/table[2]/tbody/tr/td[4]")
    for i in views_list:
        views.append(i.text)
except NoSuchElementException:#handling no such element exception
    views.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    views.append('No details available')
time.sleep(2)    
#scraping uploaded date of the video
upload_date=[]
try:
    upload_date_list=driver.find_elements(By.XPATH,"//*[@id='mw-content-text']/div[1]/table[2]/tbody/tr/td[5]")
    for i in upload_date_list:
        upload_date.append(i.text)
        
except NoSuchElementException:#handling no such element exception
    upload_date.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    upload_date.append('No details available')
    
time.sleep(2)
#preparing dataframe
df=pd.DataFrame({"Name":name, "Uploader/artist":artist, "Date of upload":upload_date,"Views(in billion)":views})
df
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)
# Opening the naukri.com
url = "https://www.naukri.com/"
driver.get(url)
time.sleep(2)
#accesing search bar
search=driver.find_element(By.XPATH,"//*[@placeholder='Enter skills / designations / companies']")
search.send_keys("Data Science")
time.sleep(1)
#clicking on search button
button=driver.find_element(By.XPATH,"//*[@class='qsbSubmit']")
button.click()
time.sleep(1)
#scrapping designation
designation=[]
try:
    designations=driver.find_elements(By.XPATH,"//*[@class='title fw500 ellipsis']")
    for i in designations:
        designation.append(i.text)
except NoSuchElementException:           #handling no such element exception
    designation.append('No details available')
time.sleep(2)
#scrapping company names
companyname=[]
try:
    companynames_list=driver.find_elements(By.XPATH,"//*[@class='subTitle ellipsis fleft']")
    for i in companynames_list:
        companyname.append(i.text)
except NoSuchElementException:        #handling no such element exception
    companyname.append('No details available')
time.sleep(2)

#skills they hire for
skill=[]
try:
    skills=driver.find_elements(By.XPATH,"//*[@class='job-description fs12 grey-text']")
    for i in skills:
        skill.append(i.text)
except NoSuchElementException:        #handling no such element exception
    skill.append('No details available')
time.sleep(2)  
#creating dataframe
df=pd.DataFrame({'Designation':designation,'Company':companyname,'Skills I hire':skill})
df
    

    

        
    

    
```
